```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load and process CSV
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Brazil']
df_selected = df[df['Country Name'].isin(selected_countries)]

# Select a few years to show
years = ['2000', '2005', '2010', '2015', '2020', '2022']
df_bar = df_selected[['Country Name'] + years].set_index('Country Name')

# Plot grouped bar chart
x = np.arange(len(years))
bar_width = 0.15

plt.figure(figsize=(12, 6))
for i, country in enumerate(df_bar.index):
    plt.bar(x + i * bar_width, df_bar.loc[country], width=bar_width, label=country)

plt.xticks(x + bar_width * 2, years)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Trends by Country (2000–2022)")
plt.legend()
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
